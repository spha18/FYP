```python
import shutil
import os
import re

PNES_path ="./pilot/PNES/"
ES_path ="./pilot/ES/"
```

```python
folders_path = os.path.join(PNES_path, 'output')
stitched_path = os.path.join(PNES_path, 'OpenPose Pose')

header = "Nose_x,Nose_y,Nose_c,Neck_x,Neck_y,Neck_c,RShoulder_x,RShoulder_y,RShoulder_c,RElbow_x,RElbow_y,RElbow_c,RWrist_x,RWrist_y,RWrist_c,LShoulder_x,LShoulder_y,LShoulder_c,LElbow_x,LElbow_y,LElbow_c,LWrist_x,LWrist_y,LWrist_c,MidHip_x,MidHip_y,MidHip_c,RHip_x,RHip_y,RHip_c,RKnee_x,RKnee_y,RKnee_c,RAnkle_x,RAnkle_y,RAnkle_c,LHip_x,LHip_y,LHip_c,LKnee_x,LKnee_y,LKnee_c,LAnkle_x,LAnkle_y,LAnkle_c,REye_x,REye_y,REye_c,LEye_x,LEye_y,LEye_c,REar_x,REar_y,REar_c,LEar_x,LEar_y,LEar_c,LBigToe_x,LBigToe_y,LBigToe_c,LSmallToe_x,LSmallToe_y,LSmallToe_c,LHeel_x,LHeel_y,LHeel_c,RBigToe_x,RBigToe_y,RBigToe_c,RSmallToe_x,RSmallToe_y,RSmallToe_c,RHeel_x,RHeel_y,RHeel_c"

pose_marker = "],\"pose_keypoints_2d\":["
face_marker = "],\"face_keypoints_2d\":["

for folder in os.listdir(folders_path):
    jsons_path = os.path.join(folders_path, folder) # where the jason files are saved
    if os.path.isdir(jsons_path):
        print(" ".join(["running",folder]))
        output_f = open(os.path.join(stitched_path, folder+".csv"), "w")
        output_f.write(header) # write header
        output_f.write("\n")
        output_f.close()
            
        sorted_jsons =  sorted(os.listdir(jsons_path))
        for json in sorted_jsons:
            if json.endswith(".json") or json.endswith(".JSON"): # read only json files
                current_json_path = os.path.join(jsons_path, json)
                data = open(current_json_path, "r").read() # read in jason file as string
                if len(data.split(pose_marker)) > 1:
                    keypoints = (data.split(pose_marker))[1].split(face_marker)[0] # extract useful bit
                    keypoints = re.sub('0,0,0,', '-1,-1,-1,', keypoints) # replace missing values with -1
                    keypoints = re.sub('0,0,0$', '-1,-1,-1', keypoints)
                else:
                    keypoints = "-1," * len(header.split(",")) # empty frame
                output_f = open(os.path.join(stitched_path, folder+".csv"), "a")
                output_f.write(keypoints.rstrip(',')) # append to csv file
                output_f.write("\n")
                output_f.close()
        print(" ".join(["finished",folder]))
```
