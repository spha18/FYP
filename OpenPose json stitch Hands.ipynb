```python
import shutil
import os
import re

PNES_path ="./pilot/PNES/"
ES_path ="./pilot/ES/"
```

```python
folders_path = os.path.join(PNES_path, 'output')
stitched_path = os.path.join(PNES_path, 'OpenPose Hands')

LH_header = "LHand_0_x,LHand_0_y,LHand_0_c,LHand_1_x,LHand_1_y,LHand_1_c,LHand_2_x,LHand_2_y,LHand_2_c,LHand_3_x,LHand_3_y,LHand_3_c,LHand_4_x,LHand_4_y,LHand_4_c,LHand_5_x,LHand_5_y,LHand_5_c,LHand_6_x,LHand_6_y,LHand_6_c,LHand_7_x,LHand_7_y,LHand_7_c,LHand_8_x,LHand_8_y,LHand_8_c,LHand_9_x,LHand_9_y,LHand_9_c,LHand_10_x,LHand_10_y,LHand_10_c,LHand_11_x,LHand_11_y,LHand_11_c,LHand_12_x,LHand_12_y,LHand_12_c,LHand_13_x,LHand_13_y,LHand_13_c,LHand_14_x,LHand_14_y,LHand_14_c,LHand_15_x,LHand_15_y,LHand_15_c,LHand_16_x,LHand_16_y,LHand_16_c,LHand_17_x,LHand_17_y,LHand_17_c,LHand_18_x,LHand_18_y,LHand_18_c,LHand_19_x,LHand_19_y,LHand_19_c,LHand_20_x,LHand_20_y,LHand_20_c"
RH_header = "RHand_0_x,RHand_0_y,RHand_0_c,RHand_1_x,RHand_1_y,RHand_1_c,RHand_2_x,RHand_2_y,RHand_2_c,RHand_3_x,RHand_3_y,RHand_3_c,RHand_4_x,RHand_4_y,RHand_4_c,RHand_5_x,RHand_5_y,RHand_5_c,RHand_6_x,RHand_6_y,RHand_6_c,RHand_7_x,RHand_7_y,RHand_7_c,RHand_8_x,RHand_8_y,RHand_8_c,RHand_9_x,RHand_9_y,RHand_9_c,RHand_10_x,RHand_10_y,RHand_10_c,RHand_11_x,RHand_11_y,RHand_11_c,RHand_12_x,RHand_12_y,RHand_12_c,RHand_13_x,RHand_13_y,RHand_13_c,RHand_14_x,RHand_14_y,RHand_14_c,RHand_15_x,RHand_15_y,RHand_15_c,RHand_16_x,RHand_16_y,RHand_16_c,RHand_17_x,RHand_17_y,RHand_17_c,RHand_18_x,RHand_18_y,RHand_18_c,RHand_19_x,RHand_19_y,RHand_19_c,RHand_20_x,RHand_20_y,RHand_20_c"

left_marker = "],\"hand_left_keypoints_2d\":["
right_marker = "],\"hand_right_keypoints_2d\":["
end_marker = "],\"pose_keypoints_3d\":["

for folder in os.listdir(folders_path):
    jsons_path = os.path.join(folders_path, folder) # where the jason files are saved
    if os.path.isdir(jsons_path):
        print(" ".join(["running",folder]))
        output_f = open(os.path.join(stitched_path, folder+".csv"), "w")
        output_f.write(LH_header+","+RH_header) # write header
        output_f.write("\n")
        output_f.close()
            
        sorted_jsons =  sorted(os.listdir(jsons_path))
        for json in sorted_jsons:
            if json.endswith(".json") or json.endswith(".JSON"): # read only json files
                current_json_path = os.path.join(jsons_path, json)
                data = open(current_json_path, "r").read() # read in jason file as string
                if len(data.split(left_marker)) > 1:
                    LH_keypoints = (data.split(left_marker))[1].split(right_marker)[0] # extract useful bit
                    LH_keypoints = re.sub('0,0,0,', '-1,-1,-1,', LH_keypoints) # replace missing values with -1
                    LH_keypoints = re.sub('0,0,0$', '-1,-1,-1', LH_keypoints)
                else:
                    LH_keypoints = "-1," * len(LH_header.split(",")) # empty frame
                if len(data.split(right_marker)) > 1:
                    RH_keypoints = (data.split(right_marker))[1].split(end_marker)[0] # extract useful bit
                    RH_keypoints = re.sub('0,0,0,', '-1,-1,-1,', RH_keypoints) # replace missing values with -1
                    RH_keypoints = re.sub('0,0,0$', '-1,-1,-1', RH_keypoints)
                else:
                    RH_keypoints = "-1," * len(RH_header.split(",")) # empty frame
                output_f = open(os.path.join(stitched_path, folder+".csv"), "a")
                output_f.write(LH_keypoints.rstrip(',')+","+RH_keypoints.rstrip(',')) # append to csv file
                output_f.write("\n")
                output_f.close()
        print(" ".join(["finished",folder]))
```
